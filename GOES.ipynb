```python
import matplotlib.pyplot as plt

from sunpy.net import Fido
from sunpy.net import attrs as a
from sunpy.time import parse_time
from sunpy.timeseries import TimeSeries

import pandas as pd
import numpy as np
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
# Rebuild the matplotlib font cache
fm._rebuild()

mpl.rcParams['font.family'] = 'Avenir'
plt.rcParams['font.size'] = 14
plt.rcParams['axes.linewidth'] = 2

plt.rcParams['text.usetex'] = True
```

```python
fn1 = 'https://umbra.nascom.nasa.gov/goes/fits/2014/go1520141024.fits'
fn2 = 'https://umbra.nascom.nasa.gov/goes/fits/2014/go1520141025.fits'
fn3 = 'https://umbra.nascom.nasa.gov/goes/fits/2014/go1520141026.fits'
```

```python
g1 = TimeSeries(fn1)
g2 = TimeSeries(fn2)
g3 = TimeSeries(fn3)
```

```python
fig, ax = plt.subplots()
g2.plot()
```

```python
g3.peek()
```

```python
i_str,i_end=27191,35750
print(g2.index[i_str],g2.index[i_end])
t = pd.Timestamp('2014-10-25T17:05:59')
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(g2.index[i_str:i_end], g2.quantity("xrsb")[i_str:i_end],label='1.0-8.0 $\AA$')
ax.plot(g2.index[i_str:i_end], g2.quantity("xrsa")[i_str:i_end],label='0.5-4.0 $\AA$')
ax.set_ylabel("Flux [W m$^{-2}$]")

ax.legend(loc=2)
ax.set_yscale('log')
ax.axvline(t,linestyle='--',color='black')
fig.autofmt_xdate()
```

```python
ind=np.concatenate((g1.index,g2.index,g3.index), axis=None)
quanb = np.concatenate((g1.quantity("xrsb"),g2.quantity("xrsb"),g3.quantity("xrsb")), axis=None)
quana = np.concatenate((g1.quantity("xrsa"),g2.quantity("xrsa"),g3.quantity("xrsa")), axis=None)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(g1.index, g1.quantity("xrsb"),label='1.0-8.0 $\AA$')
ax.plot(g1.index, g1.quantity("xrsa"),label='0.5-4.0 $\AA$')
ax.set_ylabel("Flux [W m$^{-2}$]")

ax.legend(loc=2)
ax.set_yscale('log')
fig.autofmt_xdate()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(ind, quanb,label='1.0-8.0 $\AA$')
ax.plot(ind,quana,label='0.5-4.0 $\AA$')
ax.set_ylabel("Flux [W m$^{-2}$]")

ax.legend(loc=1)
ax.set_yscale('log')
fig.autofmt_xdate()
```

```python
# set inital values
line = 1403.
mass = 28.0*1.66054e-27
kb = 1.3807e-23
c = 300
h = 6.62607e-27
flux=1.0e3 #? why this val?
```

```python
ll = np.arange(line-10,line+10,0.01)
```

```python
time= 55 # index = 0.1*time in seconds
sig = line*np.sqrt(kb*1e6*tube.tarr.t[time]/mass)/(c*1e6) # other thermal width 
n = tube.tarr.n[time]
los_v = tube.tarr.v[time].T[0]
sm_v = savgol_filter(los_v,3,1)
los_x = tube.tarr.x[time].T[0]
n_e = tube.tarr.epamu[time]*tube.tarr.rho[time]/1.67e-8 # number density 
```

```python
# process to create volume given per flux (per Maxwell)
rad=np.sqrt(flux/tube.tarr.b[time]/np.pi)
b_e=0.5*( np.roll(tube.tarr.b[time], -1) + tube.tarr.b[time] )
A1=rad
A2=np.roll(rad,-1)
#A2[tube.n-1]=A2[tube.n-2]
A2[n-1]=A2[n-2]
volume=(1./3.)*tube.tarr.dl_e[time]*(A1+A2+np.sqrt(A1*A2))*1e24
#calcualte emission measure EM
EM=volume*n_e**2
```

```python
# interpolate our GofT data
te = 10**log10T
inter = interp1d(te,log10G,kind='cubic', bounds_error=False, fill_value=-10e6) #fill outide vals with large, small number
temp = 1e6*tube.tarr.t[time]
G = inter(temp)
G[temp<22000] = -10000 # set all log10G with low temp to large, small number (st 10^G~0)
emiss = np.empty([n,len(ll)])
emiss2 = emiss
for i in range(n):
    emiss[i,:] = EM[i]*10**G[i]/np.sqrt(2*np.pi*sig[i])*np.exp(-(ll-line-line*sm_v[i]/c)**2/(2*sig[i]**2))
    emiss2[i,:] = EM[i]*10**G[i]*line/(c*1e6)*np.exp(-(ll-line-line*sm_v[i]/c)**2/(2*sig[i]**2))
half=np.int(n/2)
emiss = emiss[half:,:]
emiss2 = emiss2[half:,:]
tot_emiss = np.sum(emiss,axis=0)

#plot
fig=plt.figure(figsize=[14,8])
plt.plot(ll,tot_emiss)
plt.xlim(1401,1406)
```

## Fitting:
Taken from MCMC https://github.com/ivanzmilic/weak_field_mcmc/blob/master/Weak_field_MCMC.ipynb

```python
def ln_prior(params):
    # all parameters equally likely
    return 0.
```

```python
def ln_like(params, x, y, error):
    
    # Remember that gaussian is : p(x) = 1./(sqrt(2 pi) sigma ) exp(-x^2/2sigma^2)
    
    #What model predicts: (error == sigma)
    y_model = params[0]*(1/(params[2]*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-params[1])/params[2])**2))) + \
        params[3]*(1/(params[5]*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-params[4])/params[5])**2)))
        
    chi2 = ((y_model - y) / error)**2
    
    return np.sum((-0.5 * chi2) - 0.5*np.log(2*np.pi) - np.log(error)) # I am making it depend on errors in purpose
```

```python
def ln_post(params, x, y, error):
    
    lnprior_val = ln_prior(params)
    
    if ~np.isfinite(lnprior_val):
        return lnprior_val
    else:           
        lnlike_val = ln_like(params, x, y, error)
        return lnprior_val + lnlike_val
```

```python
# trim excess zeros in data
jj = np.where((ll > 1403) & (ll < 1405))
ll = ll[jj]
tot_emiss = tot_emiss[jj]
tot_emiss /= max(tot_emiss)
```

sampling now:

```python
ndim = 6 # because we have two parameters 
nwalkers = 20 # this is number of random points

x_data = ll
y_data = tot_emiss
errors = np.zeros(len(ll))
errors[:] = 0.001

#starting_params = np.random.uniform(0, 1, ndim*nwalkers).reshape((nwalkers, ndim))
starting_params = np.zeros((nwalkers,ndim))
starting_params[:,0] = np.random.uniform(0.8,0.12,nwalkers)
starting_params[:,1] = np.random.uniform(1403.7,1403.8,nwalkers)
starting_params[:,2] = np.random.uniform(0.04,0.06,nwalkers)
starting_params[:,3] = np.random.uniform(0.8,0.12,nwalkers)
starting_params[:,4] = np.random.uniform(1403.9,1404.04,nwalkers)
starting_params[:,5] = np.random.uniform(0.06,0.09,nwalkers)


sampler = emcee.EnsembleSampler(nwalkers, ndim, ln_post,
                                args=[x_data,y_data,errors])

num_steps = 10000
pos, prob, state = sampler.run_mcmc(starting_params, 50) # burn-in
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, num_steps)
```

```python
A1 = np.percentile(params[:,0],50.)
A2 = np.percentile(params[:,3],50.)
sig1 = np.percentile(params[:,2],50.)
sig2 = np.percentile(params[:,5],50.)
```

```python
def oneGauss(x, amp1,cen1,sigma1):
    return amp1*(1/(sigma1*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-cen1)/sigma1)**2)))

def twoGauss(x, amp1,cen1,sigma1, amp2,cen2,sigma2):
    return amp1*(1/(sigma1*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-cen1)/sigma1)**2))) + \
            amp2*(1/(sigma2*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-cen2)/sigma2)**2)))
```

```python
pars_1 = [A1,mu1,sig1]
pars_2 = [A2,mu2,sig2]
popt_2gauss = pars_1+pars_2
gauss_peak_1 = oneGauss(ll, *pars_1)
gauss_peak_2 = oneGauss(ll, *pars_2)
residual_2gauss = tot_emiss - (twoGauss(ll, *popt_2gauss))
```
